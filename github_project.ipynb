```python
to_impute_by_mean = df.loc[:, ['Total Revenue','Gross Profit',
       'Premium Revenue','Premium Cost Revenue','Premium Gross Profit',
       'Ad Revenue','Ad Cost of revenue','Ad gross Profit', 'MAUs',
       'Premium MAUs','Ad MAUs','Sales and Marketing Cost',
       'Research and Development Cost','Genreal and Adminstraive Cost']]
### selecting columns to impute the missing values by mean
```

```python
import numpy as np
```

```python
np.mean(df['Total Revenue'])
```

```python
for i in to_impute_by_mean.columns:## replacing the missing values with mean
    df[i].fillna(df[i].mean(), inplace = True)
```
